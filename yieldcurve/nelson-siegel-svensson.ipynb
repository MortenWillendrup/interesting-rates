```python
import pandas
import urllib.request
```

```python
df = pandas.read_csv(urllib.request.urlopen("http://www.bundesbank.de/cae/servlet/StatisticDownload?tsId=BBK01.WZ9801&tsId=BBK01.WZ9802&tsId=BBK01.WZ9803&tsId=BBK01.WZ9804&tsId=BBK01.WZ9805&tsId=BBK01.WZ9806&its_csvFormat=en&its_fileFormat=csv&mode=its"), index_col=0)
```

```python
mapping = dict((df.iloc[0].dropna().apply(lambda x: x.split(" / ")[1].split()[1])).items())
df.columns = [mapping.get(x,x) for x in df.columns]
df = df.iloc[4:]
```
